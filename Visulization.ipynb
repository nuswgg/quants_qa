```python
one_stock_date = QA.QA_fetch_stock_day_adv('000001','2020-05-01','2020-08-22')
df_one_stock = one_stock_date.data
df_one_stock.head()
df = df_one_stock.droplevel(level=1,axis=0)
```

```python
# 我们初始化2个部分
kline=Kline(width=1360, height=700, page_title='000001')
bar = Bar()
```

```python
import numpy as np
import pandas as pd

# 做横轴的处理
datetime = np.array(df.index.map(str))
```

```python
datetime
```

```python
ohlc = np.array(df.loc[:, ['open', 'close', 'low', 'high']])
vol = np.array(df.volume)
```

```python
kline.add('000001', datetime, ohlc, mark_point=[
      "max", "min"], is_datazoom_show=False, datazoom_orient='horizontal')

bar.add('000001', datetime, vol,
        is_datazoom_show=True,
        datazoom_xaxis_index=[0, 1])
```

```python
grid = Grid(width=1360, height=700, page_title='QUANTAXIS')
grid.add(bar, grid_top="80%")
grid.add(kline, grid_bottom="30%")
```

```python
grid.render('000001_plot.html')
```
