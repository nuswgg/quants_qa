## 定义函数

```python
def get_latest_holding(code,dt,stock_code_name):
    df = pro.fund_portfolio(ts_code=code) 
    df['code'] = df['symbol'].map(lambda x: x.replace('.SH','').replace('.SZ','').replace('.BJ',''))
    latest_data = df.query('end_date == @dt').sort_values(by='stk_mkv_ratio',ascending = 0)
    latest_holding = latest_data.merge(stock_code_name,how='inner',left_on='code',right_on='symbol')
    return latest_holding.filter(['ts_code','code','mkv','amount','stk_mkv_ratio','name'])
```
