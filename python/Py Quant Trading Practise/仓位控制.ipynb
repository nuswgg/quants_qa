```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.style.use('seaborn')
plt.rcParams['figure.figsize'] = [6,4]
plt.rcParams['figure.dpi'] = 150
```

假设一共进行200次交易

```python
n = pd.Series(np.random.randint(0,2,size = 200))
```

```python
n.replace(0,-1, inplace = True)
n.head()
```

```python
sns.countplot(x=n)
plt.show()
```

不设仓位控制  不加止盈止损  全部一把梭

```python
x1 = [10000]
for i in range(200):
    a= x1[-1]*(1+n[i])
    x1.append(a)
print(x1[-1])
```

```python
pd.Series(x1).plot(title = 'Total Asset')
plt.show()
```

每次只用25%的仓位  止盈止损全部为10%

```python
x2 = [10000]
for i in range(200):
    a = x2[-1]*0.25*(1+n[i]*0.1)+x2[-1]*0.75
    x2.append(a)
print(x2[-1])
```

```python
pd.Series(x2).plot(title='Total Asset')
plt.show()
```

仓位控制在10%  
止盈20% 止损10%

```python
x3 = [10000]
for i in range(200):
    if n[i] == 1:
        a = x3[-1]*0.1*(1+n[i]*0.2) + x3[-1]*0.9
    else:
        a = x3[-1]*0.1*(1+n[i]*0.1) + x3[-1]*0.9 
    x3.append(a)
print(x3[-1])
```

```python
pd.Series(x3).plot(title = 'Total Asset')
plt.show()
```
